Write a function named plot_variable_pairs that accepts a dataframe as input and plots all of the pairwise relationships along with the regression line for each pair.


```python
df = wrangling.get_zillow()
```

```python
df.head()
```

```python
def plot_variable_pairs(df, target, cont_cols, sample_size=500):
    '''
    Plots continuous versus continuous variable with regression line.
    '''
    for col in cont_cols:
        
        sns.lmplot(x=col, y=target, data=df.sample(sample_size),
                   robust=True, scatter_kws = {'color':'tomato'},
                   line_kws = {'color':'rebeccapurple'})
        plt.show()
```

```python
cont_cols = ['square_ft', 'taxamount', 'built']
```

```python
cat_cols = ['bedrooms', 'bathrooms', 'fips']
```

```python
plot_variable_pairs(df, 'tax_value', cont_cols)
```

Write a function named plot_categorical_and_continuous_vars that accepts your dataframe and the name of the columns that hold the continuous and categorical features and outputs 3 different plots for visualizing a categorical variable and a continuous variable.



```python
def cat_and_cont(df, target, cat_cols, quant_y_lim=0.85, sample_size=500):
    '''
    This function takes in a DataFrame, 
    target column and list of categorical columns to plot.
    '''
    
    for col in cat_cols:

        sns.swarmplot(x=col, y=target, data=df.sample(sample_size))
        plt.ylim(top=(df[target].quantile(q=quant_y_lim)), bottom=0)
        
        plt.show()
        
        sns.stripplot(x=col, y=target, data=df.sample(sample_size))
        plt.ylim(top=(df[target].quantile(q=quant_y_lim)), bottom=0)
        
        plt.show()
```

```python
cat_and_cont(df, 'tax_value', cat_cols)
```
